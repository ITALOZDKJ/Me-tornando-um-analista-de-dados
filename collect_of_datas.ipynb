# How to collect data of actions with Yahoo Finance

## Yahoo finance

```python
import yfinance as yf
from pandas import json_normalize
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
ticker = yf.Ticker('ABEV3.SA')
```

```python
df = ticker.history(period='1y', start='2016-01-01', end='2020-01-01')
```

```python
df['Close'].plot()
```

```python
decomposition = seasonal_decompose(df['Close'], period=30, extrapolate_trend=30)
```

```python
decomposition.plot()
```
