```python
df = df.dropna()
```

```python
df[['price_in_dollar','price_detail__currency', 'price_in_indian_rupee']]
```

```python
df['price_detail__amount'] = df['price_detail__amount']/74.05
```

```python
df['price_detail__amount'] = round(df['price_detail__amount'],2)
```

```python
df.rename(columns={'price_detail__amount': 'price_in_dollar'},inplace=True)
```

```python
df.rename(columns={'price_detail__price_string': 'price_in_indian_rupee'},inplace=True)
```

```python
df['price_in_indian_rupee'] = int(df['price_in_indian_rupee'])
```

```python
em_rupia = list(df['price_in_indian_rupee'])
for c in range(len(em_rupia)):
    em_rupia[c] = em_rupia[c].replace(',','.')
    em_rupia[c] = float(em_rupia[c][1:])
em_rupia
```

```python
df['price_in_indian_rupee'] = em_rupia
```
