```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
asset = yf.download('^BVSP', start= '1995-01-01')[['Close']]
```

```python
df = asset.copy()
df['Ano'] = df.index.year
df['Dia do ano'] = df.index.dayofyear
```

```python
def features_initial(x):
    x.reset_index(inplace=True)
    x['Day'] = x['Date'].dt.day
    x['Month'] = x['Date'].dt.month
    x['Year'] = x['Date'].dt.year
    x['Day-Month'] = x['Date'].dt.strftime('%m') +'-' + x['Date'].dt.strftime('%d') 
    x['Rentability%'] = round((x['Close']/x['Close'].shift()) * 100 -100, 2)
    x['Rentability pure'] = round(x['Close'] - x['Close'].shift(),2)
```

```python
features_initial(df)
```

```python
tabela = df.pivot(index=['Dia do ano'], columns='Year', values='Close')
```

```python
tab = tabela.fillna(method='bfill')
```

```python
tab = (tab/tab.iloc[0])-1
```

```python
anos = [1998,2002,2006,2010,2014,2018,2022]
```

```python
fig = px.line(
    tab[anos],
    height = 600,
    width = 800,
    template = 'plotly_dark',
    title = 'Performance ibov em anos eleitorais',
    labels = {'values': 'retorno'}
)

fig.add_vline(x=pd.to_datetime('2022-07-01').dayofyear)
fig.add_vline(x=pd.to_datetime('2022-12-31').dayofyear)
fig.layout.yaxis.tickformat = '.2%'

fig
```
